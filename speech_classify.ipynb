```python
import pandas as pd
import os
```

```python
path = "./Five/"
data = []
for file in os.listdir(path):
    print(file)
    df = pd.read_csv(path+file)
    print(df.shape)
    row = df.to_numpy().flatten()
    print(row.shape)
    data.append(row)
df_five= pd.DataFrame(data,columns=range(1292))
df_five.head()
```

```python
path = "./Happy/"
data = []
for file in os.listdir(path):
    print(file)
    df = pd.read_csv(path+file)
    print(df.shape)
    row = df.to_numpy().flatten()
    print(row.shape)
    data.append(row)
df_happy= pd.DataFrame(data,columns=range(1292))
df_happy.head()
```

```python
df_data = df_five
df_data = df_data.append(df_happy)
df_data.shape
```

```python
df_data.fillna(0,inplace=True)
df_data.head()
```
